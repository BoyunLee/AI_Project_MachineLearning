```python
da = data["GT"]
numeric_data = da.apply(pd.to_numeric, errors='coerce')
q1, q3 = np.percentile(numeric_data,[25, 75])
iqr = q3 - q1
upper_bound = q3 + (iqr * 1.5)
lower_bound = q1 - (iqr * 1.5)
c = (numeric_data > upper_bound)
outlier_count = np.count_nonzero(c)
print("이상치의 개수:", outlier_count)
```
